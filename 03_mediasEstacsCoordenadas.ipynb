```python
#con este archivo unimos las medias totales de los contaminantes (2015-2021) con el catálogo de estaciones para asignar las mediciones a cada una de ellas y a la vez
#se obtiene un archivo final con las coordenas geográficas de cada estación. 
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
path= '/content/drive/MyDrive/UPIITA/CONTAMYMETEOROLOGIA/contaminantes/'
localPath= 'stdpormes/'
```

```python

df_medias=pd.read_csv(path+localPath+'stdContamTotal.csv')
df_medias.head()
```

```python
df_medias.shape
```

UNIMOS CON LA TABLA DE ESTACIONES

```python
df_catEstac= pd.read_csv(path+'catalogos/cat_estacion.csv', encoding='latin-1',skiprows=1)
df_catEstac.head()
```

```python
df_merged = pd.merge(df_medias, df_catEstac, left_on='id_station', right_on='cve_estac')
df_merged.head()
```

```python
df_merged.shape
```

ELIMINAMOS COLUMNAS NO NECESARIAS

```python
del df_merged['id_station_x']
del df_merged['id_station_y']
```

Guardamos el archivo ya final con las mediciones promedio y
 con la ubicación de las estaciones

```python
df_merged.to_csv(path + localPath + 'stdContaminantesCoordenadas.csv',encoding='latin-1',index=False)#para que no se agregue el índice de cada archivo
```
